## Demonstrating Simulation

```python
# Defining a 5x5 robot, create simulation and show initial beliefs
R = 'r'
G = 'g'
grid = [
    [R,G,G,G,R],
    [G,G,R,G,R],
    [G,R,G,G,G],
    [R,R,G,R,G],
    [R,G,R,G,R],
]
blur = 0.05
p_hit = 200.0
simulation = sim.Simulation(grid, blur, p_hit)
simulation.show_beliefs()
```

```python
def show_rounded_beliefs(beliefs):
    '''
    Function to display probability of beliefs.
    '''
    for row in beliefs:
        for belief in row:
            print("{:0.3f}".format(belief), end="  ")
        print()
```

```python
show_rounded_beliefs(simulation.beliefs)
```
