```python
def robot_in_rectangle_world():
    from simulate import Simulation
    import simulate as sim
    import helpers
    reload(localizer)
    reload(sim)
    reload(helpers)

    R = 'r'
    G = 'g'

    grid = [ 
        [R,G,G,G,R,R,R],
        [G,G,R,G,R,G,R],
        [G,R,G,G,G,G,R],
        [R,R,G,R,G,G,G],
    ]

    blur = 0.0001
    p_hit = 100.0
    simulation = sim.Simulation(grid, blur, p_hit)
    
    #Uncomment the following lines to see grid of initial belief probabilities
    print("Initial Beliefs: ")
    simulation.show_beliefs()
    show_rounded_beliefs(simulation.beliefs)    
    initial_pos = simulation.true_pose
    print("Initial position: ", initial_pos)
    print("\n")
    
    for i in range(10):
        simulation = sim.Simulation(grid, blur, p_hit)
        simulation.run(1)
   
    #Uncomment the following lines to see grid of final belief probabilities
    print("Final Beliefs: ")
    simulation.show_beliefs()
    show_rounded_beliefs(simulation.beliefs)
    
    final_pos = simulation.true_pose
    print("True final position: ", final_pos)
    
    #Return the final position sensed by the robot, and whether it is confident of that. 
    if_knows, best_estimate = helpers.is_robot_localized(simulation.beliefs)
    if(if_knows != None):
        print("Confident: ")
    else:
        print("Not Confident: ")
    return best_estimate        
```

```python
print(robot_in_rectangle_world())
```
